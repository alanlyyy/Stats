# What is the probability of drawing first success from N trials?

```python
number_of_trials = 100000
p_success = 0.474 #pblack = pred = p_even = p_odd
p_zero = 0.026 # p_zero = p_zero_zero
```

```python
z =np.random.geometric(p_success, number_of_trials)
#print(z)
```

- Plot Histogram of Roulette rolls

```python
plt.figure(figsize=(15,15))
histogram = plt.hist(z, bins=10, range=(0,10), density=True, histtype='step')
plt.title('Number of Spins before getting first success: Black/Red, Even/Odd')
plt.xlabel('Number of spins to reach first success')
plt.ylabel('Density')
```

```python
histogram[0]
```

- The likelihood of the event from getting the first success from N spins decreases as the number of trials increases. 

        For example, the likelihood for the event to occur of getting a single success on spin number 10 is less than 1%.

- What is the probability of getting the first success from 3 spins or greater?

```python
histogram[0][3:].sum()
```

# What is the likelihood of event occuring in 10 trials?

```python
number_of_trials = 10
```

```python
#1000 tests of 10 trials
binomial = np.random.binomial(number_of_trials,p_success,1000)
```

```python
plt.figure(figsize=(15,15))
histogram = plt.hist(binomial, bins=10, range=(0,10),density=True, histtype='step')
plt.title('Event Occurs X number of times in 10 trials: Black/Red, Even/Odd')
plt.xlabel('Number of events occuring')
plt.ylabel('Density')
```

The probability of getting 4 reds in 10 trials is approximately 15%.

- What is the probability of getting 4 - 8 reds in 10 trials? 

```python
histogram[0][4:9].sum()
```

- There is a 79% likelihood of getting 4-8 reds in 10 trials.
